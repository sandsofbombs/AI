```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('pokemon.csv')

# Preview column names
print("Columns in dataset:", df.columns)

# Check if 'type1' and 'speed' exist
if 'type1' in df.columns and 'speed' in df.columns:
    
    # Group by type1 and calculate average speed
    type_speed_stats = df.groupby('type1')['speed'].agg(['mean', 'median', 'count']).sort_values(by='mean', ascending=False)
    print("\nAverage Speed by Primary Type (type1):\n")
    print(type_speed_stats)

    # Plot average speed by type
    plt.figure(figsize=(12, 6))
    sns.barplot(data=type_speed_stats.reset_index(), x='type1', y='mean', palette='viridis')
    plt.xticks(rotation=45)
    plt.title('Average Speed by Primary Pokmon Type')
    plt.xlabel('Primary Type')
    plt.ylabel('Average Speed')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

else:
    print("Your dataset must contain 'type1' and 'speed' columns to analyze element vs. speed.")
```

```python
# 
#  PART 1: Load the Dataset     
# 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('pokemon.csv')

# Display all column names for reference
print("Columns in dataset:", df.columns)
```

```python
# 
#  PART 2: Check for Required Columns     
# 

# Required columns for this analysis
required_columns = ['type1', 'speed']
missing = [col for col in required_columns if col not in df.columns]

if missing:
    print(f" Missing required columns: {missing}")
else:
    print(" All required columns are present.")
```

```python
# 
#  PART 3: Calculate Speed Stats by Type       
# 

# Group by primary type and compute speed statistics
type_speed_stats = (
    df.groupby('type1')['speed']
    .agg(['mean', 'median', 'count'])
    .sort_values(by='mean', ascending=False)
)

# Print out the table
print("\n Speed Statistics by Primary Pokmon Type:\n")
print(type_speed_stats)
```

```python
# 
#  PART 4: Visualize Speed by Type              
# 

# Reset index for bar plot
type_speed_stats_plot = type_speed_stats.reset_index()

# Create a figure with two subplots: bar plot and box plot
plt.figure(figsize=(16, 6))

#  Bar Plot: Average Speed by Type
plt.subplot(1, 2, 1)
sns.barplot(data=type_speed_stats_plot, x='type1', y='mean', palette='viridis')
plt.title('Average Speed by Primary Pokmon Type')
plt.xlabel('Primary Type')
plt.ylabel('Average Speed')
plt.xticks(rotation=45)
plt.grid(True)

#  Box Plot: Speed Distribution by Type
plt.subplot(1, 2, 2)
sns.boxplot(data=df, x='type1', y='speed', palette='coolwarm')
plt.title('Speed Distribution by Primary Pokmon Type')
plt.xlabel('Primary Type')
plt.ylabel('Speed')
plt.xticks(rotation=45)
plt.grid(True)

# Final layout
plt.tight_layout()
plt.show()

```
