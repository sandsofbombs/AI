# Does it change the potential net income if I choose either CSU or Colorado School of Mines, and if so by how much?

To compare potential net income between CSU and Colorado School of Mines, you need to assume salaries at different career stages, a flat tax rate (e.g., 25%), total cost of attendance, and potential student loan repayment. You should also assume a similar major or career path, consistent employment, and optionally consider salary growth, inflation, and cost of living.


```python
import pandas as pd

# Define salary data
data = {
    "Years_After_Graduation": [0, 5, 10],
    "CSU_Gross_Salary": [52000, 54200, 67761],
    "Mines_Gross_Salary": [76400, 79300, 90500]
}

# Create DataFrame
df = pd.DataFrame(data)

# Estimate tax rate
tax_rate = 0.25  # 25% effective tax rate (federal + state + payroll)

# Calculate net income (after tax)
df["CSU_Net_Income"] = df["CSU_Gross_Salary"] * (1 - tax_rate)
df["Mines_Net_Income"] = df["Mines_Gross_Salary"] * (1 - tax_rate)

# Add difference columns
df["Net_Income_Difference"] = df["Mines_Net_Income"] - df["CSU_Net_Income"]
df["Percent_Net_Difference"] = (df["Net_Income_Difference"] / df["CSU_Net_Income"]) * 100

# Display the table
print(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Define salary data
data = {
    "Years_After_Graduation": [0, 5, 10],
    "CSU_Gross_Salary": [52000, 54200, 67761],
    "Mines_Gross_Salary": [76400, 79300, 90500]
}

# Create DataFrame
df = pd.DataFrame(data)

# Tax assumption
tax_rate = 0.25

# Calculate net income
df["CSU_Net_Income"] = df["CSU_Gross_Salary"] * (1 - tax_rate)
df["Mines_Net_Income"] = df["Mines_Gross_Salary"] * (1 - tax_rate)

# Bar chart setup
bar_width = 0.35
index = np.arange(len(df))

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(index, df["CSU_Net_Income"], bar_width, label='CSU', color='green')
plt.bar(index + bar_width, df["Mines_Net_Income"], bar_width, label='Mines', color='blue')

# Labels and titles
plt.xlabel("Years After Graduation")
plt.ylabel("Net Income (USD)")
plt.title("Net Income Comparison: CSU vs. Colorado School of Mines")
plt.xticks(index + bar_width / 2, df["Years_After_Graduation"])
plt.legend()
plt.grid(True, axis='y', linestyle='--', alpha=0.7)

# Annotate bars with values
for i in range(len(df)):
    plt.text(index[i], df["CSU_Net_Income"][i] + 1000, f"${df['CSU_Net_Income'][i]:,.0f}", ha='center', color='black')
    plt.text(index[i] + bar_width, df["Mines_Net_Income"][i] + 1000, f"${df['Mines_Net_Income'][i]:,.0f}", ha='center', color='black')

# Show plot
plt.tight_layout()
plt.show()
```

Yes, graduates from Colorado School of Mines earn about $18,300 more per year than CSU graduates right after college, which is roughly a 47% increase in net income.
