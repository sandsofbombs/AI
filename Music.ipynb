```python
import matplotlib.pyplot as plt
import seaborn as sns

# Example data for top songs and their play counts in 2025
songs = [
    "Bad Bunny - Tití Me Preguntó",
    "Drake - Search & Rescue",
    "Taylor Swift - Anti-Hero",
    "BTS - Dynamite",
    "Ariana Grande - Positions",
    "The Weeknd - Save Your Tears",
    "Lil Baby - In a Minute",
    "Karol G - Provenza"
]

play_counts = [
    350_000_000,  # Bad Bunny
    300_000_000,  # Drake
    280_000_000,  # Taylor Swift
    250_000_000,  # BTS
    220_000_000,  # Ariana Grande
    210_000_000,  # The Weeknd
    200_000_000,  # Lil Baby
    180_000_000   # Karol G
]

# Set the plot style
sns.set(style="whitegrid")

# Create a bar plot
plt.figure(figsize=(10, 6))
bar_plot = sns.barplot(x=play_counts, y=songs, palette="viridis")

# Adding title and labels
plt.title("Top Songs of 2025 by Play Count", fontsize=16)
plt.xlabel("Number of Plays (in billions)", fontsize=12)
plt.ylabel("Songs", fontsize=12)

# Add text annotations showing the play count on top of each bar
for index, value in enumerate(play_counts):
    bar_plot.text(value + 5_000_000, index, f'{value/1_000_000:.1f}M', va='center', fontsize=11)

# Show the plot
plt.tight_layout()
plt.show()
```
